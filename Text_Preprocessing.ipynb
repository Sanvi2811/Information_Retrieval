```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize # Text Tokenization process starts
```

```python
nltk.download('punkt_tab')
```

```python
text = "This is my first practical implementation. And I am loving it"
```

```python
sentences = sent_tokenize(text)
```

```python
words = word_tokenize(text)
```

```python
print("Sentences",sentences)
print("Words", words)
```

```python
from nltk.corpus import stopwords # Stop_Word removal process starts
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
```

```python
filtered = [w for w in words if w.lower() not in stop_words and w.isalnum()]
```

```python
print("Filtered", filtered)
```

```python
from nltk.stem import PorterStemmer
 # Stemming starts here we use an Algorithm named PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
stemmed = [stemmer.stem(w) for w in filtered]
```

```python
print("Stemmer",stemmed)
```

#### Here the pipeline that is used is Given as below:
### Text Tokenization -> Stop_word Removal -> Stemming
### Text Tokenization ->  breaking sentences/para into small small pieces
### Stop_word Removal ->  removing common/low_information words (the,is,at)
### Stemming -> chopping words into their root form using fixed rules (eg ing)
